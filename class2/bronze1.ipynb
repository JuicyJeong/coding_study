```python
'''
1546번
세준이는 기말고사를 망쳤다. 세준이는 점수를 조작해서 집에 가져가기로 했다. 
일단 세준이는 자기 점수 중에 최댓값을 골랐다. 이 값을 M이라고 한다. 그리고 나서 모든 점수를 점수/M*100으로 고쳤다.

예를 들어, 세준이의 최고점이 70이고, 수학점수가 50이었으면 수학점수는 50/70*100이 되어 71.43점이 된다.
세준이의 성적을 위의 방법대로 새로 계산했을 때, 새로운 평균을 구하는 프로그램을 작성하시오.



첫째 줄에 시험 본 과목의 개수 N이 주어진다. 이 값은 1000보다 작거나 같다. 
둘째 줄에 세준이의 현재 성적이 주어진다. 이 값은 100보다 작거나 같은 음이 아닌 정수이고, 적어도 하나의 값은 0보다 크다.


첫째 줄에 새로운 평균을 출력한다. 실제 정답과 출력값의 절대오차 또는 상대오차가 10-2 이하이면 정답이다.
'''

##########################YOUR CODE##################################
'''
1. 리스트의 길이를 저장
2. 리스트로 입력을 받음
3. 리스트에서 맥스값을 반환
4. 점수/맥스값*100을 새로운 리스트에 추가
5. 갱신된 리스트 전체sum하고 len으로 나누기
 '''

list_len = int(input())
midterm_score = list(map(int, input().split()))

# list_len = 4
# midterm_score = [1, 100, 100,100]

# print(list_len)
# print(midterm_score)

max_score = max(midterm_score)
# print(max_score)

sum = 0
for i in range(0,list_len):
    midterm_score[i] = midterm_score[i]/max_score*100
    sum += midterm_score[i]
    # print(midterm_score[i])

res = sum/list_len
# print(midterm_score)
print(res)


```

```python
'''
2609번
두 개의 자연수를 입력받아 최대 공약수와 최소 공배수를 출력하는 프로그램을 작성하시오.

첫째 줄에는 두 개의 자연수가 주어진다. 이 둘은 10,000이하의 자연수이며 사이에 한 칸의 공백이 주어진다.


첫째 줄에는 입력으로 주어진 두 수의 최대공약수를, 둘째 줄에는 입력으로 주어진 두 수의 최소 공배수를 출력한다.
'''

##########################YOUR CODE##################################
'''
최대공약수란? 두 수의 최대공약수는 두 수가 서로 공통으로 가지고 있는 약수 중 가장 큰 수입니다
최소공배수란? 두 수의 최소공배수는 두 수에 서로 공통으로 존재하는 배수 중 가장 작은 수를 뜻합니다

1. 두 입력을 받음
2. 최대공약수 만들기:
    큰 수까지 반복으로 돌면서:
        나머지가 둘다 0일 경우 가장 큰 수 업데이트
3. 최소공배수 만들기:
    두 수에 대한 지정리스트를 초기화
    while true:
        두 수에 while 카운터가 늘어나면서 곱하는 값들이 각각의 리스트에 어팬드
        두 리스트를 비교해서 교집합이 있으면 그것이 최소공배수
    
    그냥 입력된 두 값 곱한거에 최대공약수로 나누면 그게 최소공배수더라...
'''

def GCD(a , b):
    res = 1
    big_num = max(a, b)
    for i in range(1,big_num+1):
        if a%i == 0 and b%i ==0:
            res = i

    return  res

    

def LCM(a, b):
    #****************기존의 접근 *********************
    # a_list = []
    # b_list = []
    # counter = 1
    # while True:
    #     a_list.append(a*counter)
    #     b_list.append(b*counter)
    #     a_set = set(a_list)
    #     b_set = set(b_list)

    #     CM = a_set.intersection(b_set)
        


    #     if len(CM) > 0:
    #         # print(min(CM))
    #         res = min(CM)
            
    #         break

    #     counter += 1
    #****************기존의 접근 *********************

    #**********최대공약수를 활용한 관계식 이용 *********
    gcd = GCD(a, b)
    res = a * b // gcd
    return res

    

a ,b = map(int, input().split())

res_gcd = GCD(a, b)
res_lcm = LCM(a ,b)


print(res_gcd)
print(res_lcm)
```
